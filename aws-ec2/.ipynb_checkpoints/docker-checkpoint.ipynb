<h1>Sagemaker Processing</h1>

```python
import boto3

region = boto3.session.Session().region_name

account_id = boto3.client('sts').get_caller_identity().get('Account')
ecr_repository = 'ds-knowledge-exchange'
tag = ':latest'
uri_suffix = 'amazonaws.com'

processing_repository_uri = '{}.dkr.ecr.{}.{}/{}'.format(account_id, region, uri_suffix, ecr_repository + tag)
```

```python
#import sagemaker
from sagemaker.processing import ScriptProcessor
from sagemaker import get_execution_role

#role = 'arn:aws:iam::694314498149:role/SagemakerExecutionRole-AdidasCltv-Restricted'
role = get_execution_role()

script_processor = ScriptProcessor(command=['python3'],
                image_uri=processing_repository_uri,
                role=role,
                instance_count=1,
                volume_size_in_gb=5,
                instance_type='ml.t3.medium')
```

```python
from sagemaker.processing import ProcessingInput, ProcessingOutput

input_data = 's3://ds-knowledge-exchange/aws-ec2/input'
output_data = 's3://ds-knowledge-exchange/aws-ec2/output'

script_processor.run(code='python/housing_pred.py',
                      inputs=[ProcessingInput(source=input_data,
                                              destination='/opt/ml/processing/input',
                                              s3_input_mode = 'File')],
                      outputs=[ProcessingOutput(destination=output_data,
                                                source='/opt/ml/processing/output',
                                                s3_upload_mode = 'Continuous')],
                      arguments=['-p', '/opt/ml/processing/input/housing.csv',
                                 '-a', 'linear',
                                 '-v', '1'
                                ]
                     )
script_processor_job_description = script_processor.jobs[-1].describe()
print(script_processor_job_description)
```
